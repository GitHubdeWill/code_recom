```python
import sys
import csv
from pprint import pprint
import os
import shutil

DATA_PATH = "/home/wtemp/585/dataset/"
STAR_PATH = "/home/wtemp/585/star_python/"
```

```python
csv.field_size_limit(sys.maxsize)
rows = []
with open('dataset/TopStaredRepositories.csv', 'r') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='"')
    for row in spamreader:
        rows.append(row)
```

```python
# pprint(rows)
# print(len(rows))
dict1 = {}  # url:[star,...]
for r in rows[1:]:
    pprint(r)
    dict1[r[-1]] = [r[-3], r[-4]]
```

```python
pprint(rows[0])
```

```python
print(len(dict1))
pprint(dict1)
```

```python
count = 0
pyfiles = []
for key, value in dict1.items():
    if dict1[key][1] == "Python":
        star = str(int(float(dict1[key][0][:-1])*1000))
        path = STAR_PATH + star + "/" + key.split("/")[-1]
        folder = os.path.exists(path)
        if not folder:
            os.makedirs(path)
        pa = DATA_PATH + key.split("/")[-1]
        
        for subdir, dirs, files in os.walk(pa):
            for file in files:
                if file.split(".")[-1] == "py":
                    count += 1
                    pyfiles.append(os.path.join(subdir, file))
                    shutil.copy(subdir+"/"+file,path)
        
count
```

```python
count = 0
with open("/home/wtemp/585/allPyPrograms.txt", "w+") as pybigfile:
    for file in pyfiles:
        count += 1
        if count%1000 == 0:
            print(count)
        with open(file, "r") as currfile:
            try:
                lines = currfile.readlines()
            except:
                continue
            for l in lines:
                if l.strip() == "" or l.strip()[0] == "#":
                    continue
#                 print(l)
                pybigfile.write(l)
```
