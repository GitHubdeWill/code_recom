```python
import sys
import csv
from pprint import pprint

DATA_PATH = "/home/wtemp/585/dataset/"
```

```python
csv.field_size_limit(sys.maxsize)
rows = []
with open('dataset/TopStaredRepositories.csv', 'r') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='"')
    for row in spamreader:
        rows.append(row)
```

```python
# pprint(rows)
# print(len(rows))
dict = {}  # url:[star,...]
for r in rows[1:]:
    pprint(r)
    dict[r[-1]] = [r[-3], r[-4]]
```

```python
count = 0
for key, value in dict.items():
    if dict[key][1] == "Python":
        count += 1
        print(key, value)
count
```

```python
import subprocess
from multiprocessing import Pool

def clone_task(url):
    # Working dorectory
    print("cloning ",url)
    p = subprocess.Popen(["git","clone",url], stdout=subprocess.PIPE, cwd="/home/wtemp/585/dataset/")
    print(p.communicate())
```

```python
urls = []
for key, value in dict.items():
    urls.append(key+".git")
# pprint(urls)
```

```python
NUM_PROCESS = 12
p = Pool(processes=NUM_PROCESS)
i = p.map_async(clone_task, urls)
```
